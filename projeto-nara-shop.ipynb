```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pedidos_narashop.csv', sep=';')
df
```

```python
df['valor_pedido'] = df['quantidade'] * df['valor_unitario']
df
```

#### O gerente quer saber quais categorias geraram mais receita. 
#### Apresente o faturamento total por categoria, do maior para o menor.

```python
df_total_categoria = df.groupby('categoria')['valor_pedido'].sum().reset_index()
df_total_categoria.sort_values(by='valor_pedido', ascending=False, inplace= True)
df_total_categoria
```

```python
df_produto = df.groupby('produto')['valor_pedido'].sum().reset_index()
df_produto.sort_values(by='valor_pedido', ascending=False, inplace=True)
df_produto.head()
```

```python
array_valor_produto = np.array(df_produto['valor_pedido'])
q1 = np.percentile(array_valor_produto, 25)
q3 = np.percentile(array_valor_produto, 75)
print(q1)
print(q3)
```

```python
df_produto_25_mais = df_produto.loc[df_produto['valor_pedido'] > q3]
df_produto_25_mais
```

```python
df_produto_25_menos = df_produto.loc[df_produto['valor_pedido'] < q1]
df_produto_25_menos
```
