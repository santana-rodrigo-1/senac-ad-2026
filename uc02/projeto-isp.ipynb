01 Faixa típica de roubo de veículo

Variância e desvio padrão

```python
# Importação de bibliotecas
import pandas as pd                  # dados
import numpy as np                   # cálculos
import matplotlib.pyplot as plt      # gráficos

# Dados em array e cálculos estatísticos
array_var = np.array(df['roubo_veiculo'])
media = np.mean(array_var)
variancia= np.var(array_var)
desvio = np.std(array_var)
faixa_inf= media - 2 * desvio
faixa_sup= media + 2 * desvio

# Saída de dados
print(f'A Média é: {media}')
print(f'Variância: {variancia:.2f}')
print(f'Desvio Padrão: {desvio:.2f}')
print(f'Faixa de Valores: {faixa_inf:.2f} a {faixa_sup:.2f}')

# Gráfico de Histograma
plt.figure(figsize=(10, 4))
plt.hist(array_var, bins=10, 
         color='#87CEFA', edgecolor='#000000' ) # color='skyblue', edgecolor='black'
plt.title('Distribuição de Roubo de Veículos')
plt.xlabel('Quantidade de Roubos')   # eixo X
plt.ylabel('Frequência')             # eixo Y
plt.show()

# Grafico de Boxplot
plt.figure(figsize=(10, 4))
plt.boxplot(array_var, vert=False, showmeans=True,
            patch_artist=True, boxprops=dict(facecolor='#87CEEB'))  # facecolor='skyblue
plt.title('Distribuição de Roubo de Veículos')
plt.xlabel('Quantidade de Roubos')   # eixo X
plt.ylabel('Dados')                  # eixo Y
plt.tight_layout()
plt.show()
```

02 Covariância entre ocorrências de veículos

furto_veiculo · roubo_veiculo · recuperacao_veiculo

```python
# Covariância entre furto, roubo e recuperação de veículos
a_furto = np.array(df['furto_veiculos'])
a_roubo = np.array(df['roubo_veiculo'])
a_recup = np.array(df['recuperacao_veiculos'])

# Furto x Roubo
cov_1 = np.cov(a_furto, a_roubo)
cov_1 = cov_1[0, 1]
print(f"Covariância furto x roubo: {cov_1:.2f}")

# Furto x Recuperação
cov_2 = np.cov(a_furto, a_recup)
cov_2 = cov_2[0, 1]
print(f"Covariância furto x recuperação: {cov_2:.2f}")

# Roubo x Recuperação
cov_3 = np.cov(a_roubo, a_recup)
cov_3 = cov_3[0, 1]
print(f"Covariância roubo x recuperação: {cov_3:.2f}")

#######################################################

# OPÇÃO E GRÁFICOS COM IA:

print("\n"'OPÇÃO E GRÁFICOS COM IA'"\n")

# Covariância entre furto, roubo e recuperação de veículos
cov_1 = np.cov(a_furto, a_roubo)[0, 1]
cov_2 = np.cov(a_furto, a_recup)[0, 1]
cov_3 = np.cov(a_roubo, a_recup)[0, 1]
print(f"Covariância furto x roubo: {cov_1:.2f}")
print(f"Covariância furto x recuperação: {cov_2:.2f}")
print(f"Covariância roubo x recuperação: {cov_3:.2f}")

# Gráfico de Linhas para Furto, Roubo e Recuperação  ## IA
plt.plot(df['furto_veiculos'], label='Furto')
plt.plot(df['roubo_veiculo'], label='Roubo')
plt.plot(df['recuperacao_veiculos'], label='Recuperação')

plt.title('Furto x Roubo x Recuperação')
plt.legend()

plt.show()

# Gráfico de Linhas para Furto, Roubo e Recuperação (com melhorias) ## IA
plt.figure(figsize=(10,5))

plt.plot(df['furto_veiculos'], label='Furto')
plt.plot(df['roubo_veiculo'], label='Roubo')
plt.plot(df['recuperacao_veiculos'], label='Recuperação')

plt.title('Análise de Veículos')
plt.xlabel('Período')
plt.ylabel('Quantidade')

plt.legend()
plt.grid()

plt.show()
```

03 Correlação entre ocorrências de veículos

Força e direção da relação entre os mesmos três pares


```python
# Correlação de Pearson entre furto, roubo e recuperação de veículos

# Correlação de Pearson entre furto e roubo
corr_1 = np.corrcoef(a_furto, a_roubo)
print(f"Pearson r = {corr_1[0,1]:.4f}")

# Correlação de Pearson entre furto e recuperação
corr_2 = np.corrcoef(a_furto, a_recup)
print(f"Pearson r = {corr_2[0,1]:.4f}")

# Correlação de Pearson entre roubo e recuperação
corr_3 = np.corrcoef(a_roubo, a_recup)
print(f"Pearson r = {corr_3[0,1]:.4f}")

###########################################################################

# OPÇÃO E GRÁFICOS COM IA:

# Correlação de Pearson entre furto, roubo e recuperação de veículos
corr_furto_roubo = corr_1[0,1]
corr_furto_recup = corr_2[0,1]
corr_roubo_recup = corr_3[0,1]

# Saída de dados
print('\n')
print(f'OPÇÃO E GRÁFICOS COM IA:')
print('\n')     
print(f"Correlação de Pearson (furto x roubo): {corr_1[0,1]:.4f}")
print(f"Correlação de Pearson (furto e recuperação): {corr_2[0,1]:.4f}")
print(f"Correlação de Pearson (roubo e recuperação): {corr_3[0,1]:.4f}")



#################################

# Valores das correlações
valores_corr = [corr_furto_roubo, corr_furto_recup, corr_roubo_recup]

# Nomes das relações
relacoes = ['Furto x Roubo', 'Furto x Recuperação', 'Roubo x Recuperação']

# Gráfico de Barras Horizontal para Correlação de Pearson
plt.figure(figsize=(10, 8))
plt.barh(relacoes, valores_corr,
         color='#4682B4')  # steelblue
plt.title('Correlação de Pearson entre Variáveis')
plt.tight_layout()
plt.show()
```
